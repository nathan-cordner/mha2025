### Extract City, County, State, Country

```python
# Extract city, county, state, country

# Export US cities to file 

us_cities = ["New York City, New York, United States",
             "Philadelphia, Pennsylvania, United States",
             "Washington, District Of Columbia, United States",
             "Norfolk, Virginia, United States"]

uk_counties = ["Carmarthenshire, Wales, United Kingdom",
               "Monmouthshire, Wales, United Kingdom",
               "Nottinghamshire, England, United Kingdom",
               "Pembrokeshire, Wales, United Kingdom"]

provinces = ["South Holland, Netherlands",
             "Tahiti, French Polynesia",
             "Nordjylland, Denmark"]

city = []
county = []
state = []
country = []

places = list(df["new_residence"])

for k in places:
    cur_key = str(k).split(",")    
    
    if (len(cur_key) == 1):
        country.append(cur_key[-1].strip())
        state.append("")
        county.append("")
        city.append("")    
    elif (len(cur_key) >= 4):
        country.append(cur_key[-1].strip())
        state.append(cur_key[-2].strip())            
        county.append(cur_key[-3].strip())
        city.append(cur_key[-4].strip())
    
    # US locations 
    elif "United States" in cur_key[-1]:    
        if len(cur_key) == 2:
            country.append(cur_key[-1].strip())
            state.append(cur_key[-2].strip())
            county.append("")
            city.append("")            
        if len(cur_key) == 3:
            country.append(cur_key[-1].strip())
            state.append(cur_key[-2].strip())            
            if k.strip() in us_cities:
                county.append("")
                city.append(cur_key[-3].strip())
            else:
                county.append(cur_key[-3].strip())
                city.append("")
            
    # Canadian locations
    elif "Canada" in cur_key[-1]:        
        if len(cur_key) == 2:
            country.append(cur_key[-1].strip())
            state.append(cur_key[-2].strip())
            county.append("")
            city.append("")            
        if len(cur_key) == 3:
            country.append(cur_key[-1].strip())
            state.append(cur_key[-2].strip())            
            county.append("")
            city.append(cur_key[-3].strip())            
    # UK locations
    elif "United Kingdom" in cur_key[-1]:        
        if len(cur_key) == 2:
            country.append(cur_key[-1].strip())
            state.append(cur_key[-2].strip())
            county.append("")
            city.append("")            
        if len(cur_key) == 3:
            country.append(cur_key[-1].strip())
            state.append(cur_key[-2].strip())
            if k.strip() in uk_counties:
                county.append(cur_key[-3].strip())
                city.append("")
            else:
                county.append("")
                city.append(cur_key[-3].strip())
    elif "Mexico" in cur_key[-1]:
        # all have length 3 or 4
        country.append(cur_key[-1].strip())
        state.append(cur_key[-2].strip())            
        county.append("")
        city.append(cur_key[-3].strip())
    # all other countries
    else:
        if len(cur_key) == 2:
            country.append(cur_key[-1].strip())
            if k.strip() in provinces:
                state.append(cur_key[-2].strip())            
                county.append("")
                city.append("")
            else:
                state.append("")    
                county.append("")
                city.append(cur_key[-2].strip())          
        if len(cur_key) == 3:
            country.append(cur_key[-1].strip())
            state.append(cur_key[-2].strip())            
            county.append("")
            city.append(cur_key[-3].strip())

df["City"] = city
df["County"] = county
df["State"] = state
df["Country"] = country
            
"""
mission_city_dict = {
    "Residence": places,
    "City":  city,
    "County": county,
    "State": state,
    "Country": country
}            

city_df = pd.DataFrame(mission_city_dict)
# city_df.sort_values(by="Count", inplace=True, ascending=False)
# city_df.to_csv("city_count.csv", encoding="utf-8", index=False)

city_df.head(20)
"""
```

### Add Treemap values

```python
# Treemap data
# Category:  United States, Outside, Unknown
# Subcategory:  State, Country, Unknown

categories = []
subcategories = []

for i in range(len(df)):    
    cur_key = str(df["new_residence"].iloc[i]).split(",")  
    cur_val = cur_key[-1].strip()
    
    if "United States" in cur_val:
        categories.append("United States")
        if len(cur_key) == 1:
            subcategories.append("United States")
        else:
            subcategories.append(cur_key[-2].strip())
    elif "nan" in cur_val:
        categories.append("Unknown")
        subcategories.append("Unknown")
    else:
        categories.append("Outside US")
        subcategories.append(cur_key[-1].strip())

df["Category"] = categories
df["Subcategory"] = subcategories


df.head(20)   
    
    
```

```python
### Create Subsubcategory
### Splits Utah and Idaho into counties 

subsubcategory = []

for i in range(len(df)):
    cur_category = str(df["Subcategory"].iloc[i]).strip()
    if cur_category == "Utah" or cur_category == "Idaho":
        ### add in county 
        subcategory = str(df["County"].iloc[i]).strip()
        if not subcategory == "nan": # not empty string
            cur_category = f"{subcategory}, {cur_category}"
    subsubcategory.append(cur_category)

df["Subsubcategory"] = subsubcategory
    
# check
# subsubcategory = sorted(list(set(subsubcategory)))
#
# for x in subsubcategory:
#     print(x)
```

### Find Coordinates

```python
from geopy.geocoders import Nominatim
loc = Nominatim(user_agent="GetLoc")


# get city strings
with open ("unique_cities.txt", "r", encoding="utf-8") as f:
    cities = f.readlines()
    
coords = []

for c in cities: 
    # entering the location name
    try:
        getLoc = loc.geocode(c)
    except:
        getLoc = None
    
    print(c.strip())
 
    if getLoc:
        
        coord_str = f"{getLoc.latitude}, {getLoc.longitude}"
        coords.append(coord_str)
        # printing address
        # print("Standard:", getLoc.address)
 
        # printing latitude and longitude
        # print(f"{getLoc.latitude}, {getLoc.longitude}")
    else:
        coords.append("Unknown")
        # print("Unknown")
    # print()
    

coord_dict = {
    "City": cities,
    "Coordinates": coords
}    
    
df = pd.DataFrame(coord_dict)
df.to_csv("city_coords_python.csv", index=False)
```

### Add City Geolocation to DataFrame

```python
# read in coordinates

coord_df = pd.read_csv("city_coords_python.csv", encoding="unicode_escape")

# coord_df

```

```python
def clean_city(x):
    x = x.strip()
    if x[-1] == "n" and x[-1] == "\\":
        x = x[:-2]
    return x

coord_df["City"] = coord_df["City"].apply(clean_city)
coord_df["Coordinates"] = coord_df["Coordinates"].apply(lambda x:  x.strip())
```

```python
# Create dictionary

coords_dict = {}

for i in range(len(coord_df)):
    cur_city = str(coord_df["City"].iloc[i])
    
    # latitude, longitude
    cur_coords = str(coord_df["Coordinates"].iloc[i]).split(", ")
    
    coords_dict[cur_city] = cur_coords 
    
latitudes = []
longitudes = []

for i in range(len(df)):
    cur_residence = str(df["new_residence"].iloc[i]).strip()
    if cur_residence in coords_dict:
        latitudes.append(coords_dict[cur_residence][0])
        longitudes.append(coords_dict[cur_residence][1])
    else:
        latitudes.append("")
        longitudes.append("")
df["latitude"] = latitudes
df["longitude"] = longitudes
    
```
